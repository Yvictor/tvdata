```python
import requests
import time
import datetime as dt
import pandas as pd
```

```python
HISTOCK_UDF = 'https://histock.tw/Stock/tv/udf.asmx/'
HEADERS = {
    "User-Agent":
        "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36"
}
```

```python
def fetch_histock_udf(symbol, res, st_ts, end_ts):
    res = requests.get(f'{HISTOCK_UDF}history?symbol={symbol}'\
        f'&resolution={res}&from={st_ts}&to={end_ts}',
        headers=HEADERS)
    data = res.json()
    data['t'] = [pd.Timestamp(dt.datetime.fromtimestamp(ts)-dt.timedelta(days=1)) for ts in data['t']]
    del data['s']
    del data['nextTime']
    return data
```
